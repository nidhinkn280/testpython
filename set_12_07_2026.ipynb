```python
#set operations
set1={1,2,3,4,5}
set2={3,4,5,6}
print(set1|set2)
```

```python
#set operation-Union
set1={1,2,3}
set2={3,4,5}
print(set1.union(set2))# combines all unique elements
```

```python
#set operation-Intersection
set1={1,2,3,4}
set2={3,4,5}
print(set1.intersection(set2))
```

```python
#set operation-Union
set1={1,2,3}
set2={3,4,5}
print(set1 & set2)
```

```python
#set operation-Union-differnece
set1={1,2,3}
set2={3,4,5}
print(set1-set2)
```

```python
#set operation-Union
set1={1,2,3}
set2={3,4,5}
print(set1^set2)
```

```python
#set operation-Union
set1={1,2,3}
set2={3,4,5}
print(set1.symmetric_difference(set2))
```

```python
#issubset
set1={1,2}
set2={1,2,3,4}
print(set1.issubset(set2))
```

```python
#superset
set1={1,2,3,4}
set2={1,2}
print(set1.issuperset(set2))
```

```python
#set operation-Union
set1={1,2,3}
set2=set1.copy()
print(set2)
```
